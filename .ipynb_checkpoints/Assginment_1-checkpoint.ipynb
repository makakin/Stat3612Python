## Step 1:

Download the “`timesData.csv`” file. Select the subdata with top 200 world-ranked universities. Identify all the numeric types of variables that can be represented by floating point numbers, and then summarize them.


<b>Solution:</b>

Among the 14 variables, we are interested in `world_rank`, `university_name`, `country`, `total_score`, `teaching`, `research`, `citations` and `year`.

```python
#load data from file and preview the dataset
df = pd.read_csv('./HW_datasets/timesData.csv')
df.tail()
```

```python
#check the types of all variables
df.dtypes
```

**The data types of the following columns can be transformed from `object` to `float64` or `int64`:**
1. `world_rank`
2. `international`
3. `income`
4. `total_score`
5. `num_students`
6. `international_students`
7. `female_male_ratio`

**Let's start with `world_rank`**

```python
#check the string lengths of world_rank
df['world_rank'].str.len().unique()
```

The lengths of world rank are supposed to be of length 1, 2, 3 or 7. We need to look into those with length of 4 for anomalies.

```python
#look into world_rank with length of 4
df['world_rank'][df['world_rank'].str.len() == 4].tail()
```

```python
df['world_rank'][df['world_rank'].str.len() == 7].head()
```

Therfore, to select the top 200 universities, we could select all entries whose length of `world_rank` is equal or under 4. To change the data type of `world_rank`, we should first remove the "=" sign.

```python
#select subdata with top 200 world-ranked universities
df = df[df['world_rank'].str.len() <= 4]
#remove the equal sign and change the data type
df.loc[:,'world_rank'] = df.loc[:,'world_rank'].str.replace('=','').astype(int)
```

Other variables are adjusted based on their own patterns.

```python
df.loc[:,'num_students'] = df.loc[:,'num_students'].str.replace(',','').astype(float)
df.loc[:,'international'] = df.loc[:,'international'].replace("-", np.nan).astype(float)
df.loc[:,'international_students'] = df.loc[:,'international_students'].str.strip("%").astype(float)
df.loc[:,'income'] = df.loc[:,'income'].replace("-", np.nan).astype(float)
df.loc[:,'total_score'] = df.loc[:,'total_score'].astype(float)
temp = df['female_male_ratio'].str.split(':')
for i in range(temp.index[-1]+1):
    try:
        temp[i] = int(temp[i][0].strip())/float(temp[i][1].strip())
    except:
        continue
df.loc[:,'female_male_ratio'] = temp
df['female_male_ratio'] = df['female_male_ratio'].astype(float)
```

By now we have transformed the data type of all numerical entries.

```python
df.dtypes
```

## Step 3:

Perform data subsetting with conditions `country == "Hong Kong"` and `year > 2011`, followed by a graphical method to show the ranking changes of local universities over the years.

**Solution:**

```python
df_hk = df[(df['country'] == 'Hong Kong') & (df['year'] > 2011)]
```

```python
df_hk.head()
```

```python
def plot_ranking_hk(grouped):
    fig, ax = plt.subplots()
    fig.set_size_inches(8, 8)

    grouped.plot(x='year',ax=ax, ls='-', marker='o', grid=True)

    labels = [] #for legend
    for label, group in grouped:
        labels.append(label)
    plt.ylim([200,0])
    plt.legend(labels,loc='best', fontsize=10)

    ax.set_xticks(np.arange(2012, 2017, 1))
    ax.set_xticklabels(range(2012, 2017))

    plt.title('World Rank of Universities in Hong Kong', fontsize=16)
    plt.ylabel('World Rank', fontsize=14)
    plt.xlabel('Year', fontsize=14)

    plt.show()
```

```python
grouped = df_hk.loc[:,['world_rank','university_name','year']].groupby('university_name')
plot_ranking_hk(grouped)
```

If you choose to remove CityU from the plot

```python
df_hk = df_hk[df_hk['university_name'] != 'City University of Hong Kong']
grouped = df_hk.loc[:,['world_rank','university_name','year']].groupby('university_name')

plot_ranking_hk(grouped)
```
