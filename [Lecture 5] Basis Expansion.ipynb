```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import numpy.random as rnd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def set_font():
    font = {'weight' : 'normal',
            'size'   : 14}
    plt.rc('font', **font)
```

# 1. Nonparametric Regression
---

```python
ffun = lambda x: np.exp(-(x-3)**2)
x = np.arange(0.1, 4.1 ,0.1)
rnd.seed(2017)
y = ffun(x) + 0.1*rnd.normal(size=len(x))
```

```python
plt.figure(figsize=(10,6));
plt.plot(x, y, 'ko', x, ffun(x), 'r-');
plt.xlabel('x');
plt.ylabel('y');
set_font()
```

## 1.1. Polynomial

```python
def curve_fitting(x, y, degrees):
    p = np.polyfit(x, y, degrees)
    q = np.poly1d(p)
    return q(x)
```

```python
plt.figure(figsize=(10,6));
plt.plot(x, y, 'ko', x, ffun(x), 'r-', x, curve_fitting(x, y, 3));
plt.xlabel('x');
plt.ylabel('y');
plt.legend(['Data Points', 'ffun', 'Cubic Fitting'], loc=2)
set_font()
```

## 1.2. Fourier

```python
xs = np.arange(-2*np.pi, 2*np.pi+np.pi/100, np.pi/100)
```

```python
wave1 = np.sin(xs*3)
wave2 = np.sin(10*xs)
```

```python
plt.figure(figsize=(16,6))
plt.subplot(121)
plt.plot(xs, wave1,[-8,8], [0,0], 'k:');
plt.xlabel('xs');
plt.ylabel('wave1');

plt.subplot(122)
plt.plot(xs, wave2, [-8,8], [0,0], 'k:');
plt.xlabel('xs');
plt.ylabel('wave2');
```

```python
wave3 = 0.5 * wave1 + .25 * wave2
```

```python
plt.figure(figsize=(16,6))
plt.plot(xs, wave3, [-8,8], [0,0], 'k:');
plt.xlabel('xs');
plt.ylabel('wave 3');
plt.title('Eg complex wave')
```
